```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Define a function to interpolate points along a line between two points
def interpolate_points(p1, p2, num_points=10):
    """Interpolate points between two coordinates (p1, p2)."""
    x_values = np.linspace(p1[0], p2[0], num_points)
    y_values = np.linspace(p1[1], p2[1], num_points)
    return list(zip(x_values, y_values))

# Adjusted coordinates for the word WELCOME with tighter spacing between L, C, O, M
w_coords = interpolate_points((0, 5), (0.5, 0), 20) + \
           interpolate_points((0.5, 0), (1, 3), 10) + \
           interpolate_points((1, 3), (1.5, 0), 10) + \
           interpolate_points((1.5, 0), (2, 5), 20)

e_coords = interpolate_points((2.5, 0), (2.5, 5), 20) + \
           interpolate_points((2.5, 5), (3.5, 5), 10) + \
           interpolate_points((2.5, 2.5), (3, 2.5), 10) + \
           interpolate_points((2.5, 0), (3.5, 0), 10)

l_coords = interpolate_points((4, 0), (4, 5), 20) + \
           interpolate_points((4, 0), (5, 0), 10)

c_coords = interpolate_points((5.5, 0), (5.5, 5), 20) + \
           interpolate_points((5.5, 5), (6.5, 5), 10) + \
           interpolate_points((5.5, 0), (6.5, 0), 10)

o_coords = interpolate_points((7, 0), (7, 5), 20) + \
           interpolate_points((7, 5), (8, 5), 10) + \
           interpolate_points((8, 5), (8, 0), 20) + \
           interpolate_points((8, 0), (7, 0), 10)

m_coords = interpolate_points((8.5, 0), (8.5, 5), 20) + \
           interpolate_points((8.5, 5), (9.5, 3), 10) + \
           interpolate_points((9.5, 3), (10.5, 5), 10) + \
           interpolate_points((10.5, 5), (10.5, 0), 20)

e2_coords = interpolate_points((11, 0), (11, 5), 20) + \
            interpolate_points((11, 5), (12, 5), 10) + \
            interpolate_points((11, 2.5), (11.5, 2.5), 10) + \
            interpolate_points((11, 0), (12, 0), 10)

# Coordinates for the word TECH 26 with wider spacing between T, E, C, H
t_coords = interpolate_points((5, -2), (5, -5), 20) + \
           interpolate_points((4.5, -2), (5.5, -2), 10)  # Correct T orientation

e3_coords = interpolate_points((6, -5), (6, -2), 20) + \
            interpolate_points((6, -2), (7, -2), 10) + \
            interpolate_points((6, -3.5), (6.5, -3.5), 10) + \
            interpolate_points((6, -5), (7, -5), 10)

c2_coords = interpolate_points((7.5, -5), (7.5, -2), 20) + \
            interpolate_points((7.5, -2), (8.5, -2), 10) + \
            interpolate_points((7.5, -5), (8.5, -5), 10)

h_coords = interpolate_points((9, -5), (9, -2), 20) + \
           interpolate_points((9, -3.5), (10, -3.5), 10) + \
           interpolate_points((10, -5), (10, -2), 20)

# Fixed coordinates for the number 2 (distinct, open bottom part)
two_coords = interpolate_points((11, -3), (12, -3), 10) + \
             interpolate_points((12, -3), (12, -4), 10) + \
             interpolate_points((12, -4), (11, -5), 10) + \
             interpolate_points((11, -5), (12, -5), 10)

# Coordinates for the number 6
six_coords = interpolate_points((13, -4), (13, -5), 10) + \
             interpolate_points((13, -5), (14, -5), 10) + \
             interpolate_points((14, -5), (14, -4), 10) + \
             interpolate_points((14, -4), (13, -4), 10) + \
             interpolate_points((13, -4), (13, -3), 10)

# Combine the coordinates for all letters (WELCOME and TECH 26)
combined_coords = w_coords + e_coords + l_coords + c_coords + o_coords + m_coords + e2_coords + \
                  t_coords + e3_coords + c2_coords + h_coords + two_coords + six_coords

# Unpack combined coordinates into numpy arrays
final_x = [x for x, y in combined_coords]
final_y = [y for x, y in combined_coords]

# Convert final_x and final_y to numpy arrays
final_x = np.array(final_x)
final_y = np.array(final_y)

# Generate initial random points as numpy arrays
num_points = len(final_x)
start_x = np.random.uniform(0, 16, num_points)  # Increased range for more spacing in "WELCOME" and "TECH 26"
start_y = np.random.uniform(-7, 6, num_points)  # Adjusted Y range to accommodate "TECH 26"

# Plot setup for WELCOME and TECH 26
fig, ax = plt.subplots()
scatter = ax.scatter(start_x, start_y)
ax.set_xlim(0, 16)  # Adjusted X-axis range for the full word WELCOME and TECH 26
ax.set_ylim(-7, 6)  # Adjusted Y-axis limit to fit both words
ax.set_aspect('equal')  # Ensure equal aspect ratio
ax.set_xticks([])
ax.set_yticks([])

# Update function for the animation
def update(frame):
    # Interpolate between initial and final positions based on the current frame
    t = frame / 100  # Normalized time (0 to 1)

    # Interpolating the positions
    new_x = (1 - t) * start_x + t * final_x
    new_y = (1 - t) * start_y + t * final_y

    # Update the scatter plot with new positions
    scatter.set_offsets(np.c_[new_x, new_y])
    return scatter,

# Create the animation
ani = FuncAnimation(fig, update, frames=100, interval=50, blit=True)

# Save the animation as a GIF in Colab
ani.save("/content/welcome_tech26_final.gif", writer='pillow')

# Display the GIF inline in Colab
HTML("""
<img src="/content/welcome_tech26_final.gif" width="600">
""")
```
